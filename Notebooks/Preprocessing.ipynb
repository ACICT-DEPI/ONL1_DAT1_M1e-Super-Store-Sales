# **Data Preprocessing**

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
duplicates
```

```python
# Convert Postal Code to string (object type)
df['Postal Code'] = df['Postal Code'].astype(str)
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
# Check the data types of all columns
df.dtypes
```

```python
# Check for rows where Quantity is zero
zero_quantity = df[df['Quantity'] == 0]

# Output the number of rows with zero quantity
zero_count = zero_quantity.shape[0]
print(f'Number of rows with zero Quantity: {zero_count}')
```

```python
# drop rows where ship date is earlier than order date
df = df[df['Ship Date'] >= df['Order Date']]
df
```

```python
negative_sales = df[df['Sales'] < 0]

# Display the rows where sales are negative
print(negative_sales)
```

```python
# Filter rows where discount is not between 0 and 1
invalid_discount = df[(df['Discount'] < 0) | (df['Discount'] > 1)]

# Display the rows with invalid discount values
print(invalid_discount)
```

```python
inconsistent_profit = df[(df['Discount'] > 0.5) & (df['Profit'] > df['Sales'])]
print(inconsistent_profit)
```

# **Feature Engineering**

```python
#feature creation (profit per unit)
df['profit per unit']=df['Profit']/df['Quantity']
```

```python
df['ShippingTime'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
#drop unnecessary columns
df = df.drop(['Row ID', 'Postal Code'], axis=1)
```
