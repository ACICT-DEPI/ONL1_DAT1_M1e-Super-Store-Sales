# **Data Preprocessing**

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
duplicates
```

```python
# Convert Postal Code to string (object type)
df['Postal Code'] = df['Postal Code'].astype(str)
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
# Check the data types of all columns
df.dtypes
```

```python
# Check for rows where Quantity is zero
zero_quantity = df[df['Quantity'] == 0]

# Output the number of rows with zero quantity
zero_count = zero_quantity.shape[0]
print(f'Number of rows with zero Quantity: {zero_count}')
```

```python
# drop rows where ship date is earlier than order date
df = df[df['Ship Date'] >= df['Order Date']]
df
```

```python
negative_sales = df[df['Sales'] < 0]

# Display the rows where sales are negative
print(negative_sales)
```

```python
# Filter rows where discount is not between 0 and 1
invalid_discount = df[(df['Discount'] < 0) | (df['Discount'] > 1)]

# Display the rows with invalid discount values
print(invalid_discount)
```

```python
inconsistent_profit = df[(df['Discount'] > 0.5) & (df['Profit'] > df['Sales'])]
print(inconsistent_profit)
```

```python
#feature creation (profit per unit)
df['profit per unit']=df['Profit']/df['Quantity']
```

```python
df['ShippingTime'] = (df['Ship Date'] - df['Order Date']).dt.days
```

**Customer behavior**

```python
# Finding frequently bought together products (using a simple approach)
from itertools import combinations
from collections import Counter

# Create a list of all products in each order
order_products = df.groupby('Order ID')['Product Name'].apply(list)

# Count combinations of products in each order
product_pairs = []
for products in order_products:
    pairs = combinations(products, 2)
    product_pairs.extend(pairs)

# Count the frequency of each pair
pair_counts = Counter(product_pairs)
top_pairs = pair_counts.most_common(10)

print("Top 10 Frequently Bought Together Products:")
for pair, count in top_pairs:
    print(f"{pair[0]} and {pair[1]}: {count} times")
```

```python
# Converting Order Date to datetime (if not already done)
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Creating new columns for year, month, and quarter
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter

# Grouping by Year, Month, and Quarter for Sales and Profit trends
monthly_trends = df.groupby(['Year', 'Month']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
quarterly_trends = df.groupby(['Year', 'Quarter']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
yearly_trends = df.groupby('Year').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Visualization: Monthly sales and profit trends
plt.figure(figsize=(12, 6))
plt.plot(monthly_trends['Sales'], label='Sales', color='blue', marker='o')
plt.plot(monthly_trends['Profit'], label='Profit', color='green', marker='o')
plt.title('Monthly Sales and Profit Trends')
plt.xlabel('Time (Months)')
plt.ylabel('Amount ($)')
plt.legend()
plt.grid(True)
plt.show()

# Visualization: Quarterly sales and profit trends
plt.figure(figsize=(12, 6))
plt.plot(quarterly_trends['Sales'], label='Sales', color='blue', marker='o')
plt.plot(quarterly_trends['Profit'], label='Profit', color='green', marker='o')
plt.title('Quarterly Sales and Profit Trends')
plt.xlabel('Time (Quarters)')
plt.ylabel('Amount ($)')
plt.legend()
plt.grid(True)
plt.show()

# Visualization: Yearly sales and profit trends
plt.figure(figsize=(10, 6))
plt.bar(yearly_trends['Year'], yearly_trends['Sales'], color='skyblue', label='Sales')
plt.bar(yearly_trends['Year'], yearly_trends['Profit'], color='lightgreen', alpha=0.7, label='Profit')
plt.title('Yearly Sales and Profit Trends')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.legend()
plt.show()
```

```python
# Identify specific periods with highest sales and profit (e.g., holiday seasons, year-end)
# Finding the periods with the highest sales and profit
top_months = monthly_trends.sort_values(by='Sales', ascending=False).head(5)
top_quarters = quarterly_trends.sort_values(by='Sales', ascending=False).head(5)

# Visualization: Bar plot for months with highest sales
plt.figure(figsize=(10, 6))
plt.bar(top_months['Month'].astype(str) + '-' + top_months['Year'].astype(str), top_months['Sales'], color='orange')
plt.xlabel('Month-Year')
plt.ylabel('Sales')
plt.title('Top 5 Months with Highest Sales')
plt.show()

# Visualization: Bar plot for quarters with highest sales
plt.figure(figsize=(10, 6))
plt.bar(top_quarters['Quarter'].astype(str) + '-' + top_quarters['Year'].astype(str), top_quarters['Sales'], color='purple')
plt.xlabel('Quarter-Year')
plt.ylabel('Sales')
plt.title('Top 5 Quarters with Highest Sales')
plt.show()
```

```python
# Time-series analysis: Monthly sales and profit trends
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Resample by month and sum Sales and Profit
monthly_trends = df.resample('M', on='Order Date').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Visualization of monthly trends
plt.figure(figsize=(12, 6))
plt.plot(monthly_trends['Order Date'], monthly_trends['Sales'], label='Sales', color='blue', marker='o')
plt.plot(monthly_trends['Order Date'], monthly_trends['Profit'], label='Profit', color='green', marker='o')
plt.title('Monthly Sales and Profit Trends')
plt.xlabel('Date')
plt.ylabel('Amount ($)')
plt.legend()
plt.grid(True)
plt.show()
```

**Apriori Association Rules with Filtering**

```python
# Import required libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset (replace this with your actual dataset)
# Assuming df is your original dataset containing OrderID, ProductID, Quantity, and Profit columns

# One-hot encoding for products in each order
basket = df.groupby(['Order ID', 'Product ID'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('Order ID')
basket = basket.applymap(lambda x: 1 if x > 0 else 0)

# Apply Apriori algorithm
frequent_itemsets = apriori(basket, min_support=0.001, use_colnames=True)

# Generate association rules with a minimum confidence threshold of 0.5
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Display the first few rules
print("Generated Association Rules:")
print(rules.head())

# OPTIONAL: Add product profit classification to analyze high and low profit products
# You can create a custom classification for products based on their profit
# For example, let's classify products as 'High Profit' and 'Low Profit' based on their average profit

# Create a product profit classification (e.g., high vs low)
profit_threshold = 0  # Define threshold for profit classification
product_profit = df.groupby('ProductID').agg({'Profit': 'mean'}).reset_index()
product_profit['Profit Category'] = product_profit['Profit'].apply(lambda x: 'High Profit' if x > profit_threshold else 'Low Profit')

# Merge profit category back to the basket data
df_with_profit_category = pd.merge(df, product_profit[['Product ID', 'Profit Category']], on='Product ID', how='left')

# Filter rules based on high or low-profit products
# We want to know if high-profit items are frequently bought together with low-profit items
def filter_rules_by_profit_category(rules, df, category):
    # Find the products in the given category (high or low profit)
    products_in_category = df[df['Profit Category'] == category]['Product ID'].unique()

    # Filter rules where either antecedents or consequents contain products in the category
    filtered_rules = rules[rules['antecedents'].apply(lambda x: any(item in products_in_category for item in x)) |
                           rules['consequents'].apply(lambda x: any(item in products_in_category for item in x))]

    return filtered_rules

# Filter rules for high-profit products
high_profit_rules = filter_rules_by_profit_category(rules, df_with_profit_category, 'High Profit')
print("\nAssociation Rules for High-Profit Products:")
print(high_profit_rules.head())

# Filter rules for low-profit products
low_profit_rules = filter_rules_by_profit_category(rules, df_with_profit_category, 'Low Profit')
print("\nAssociation Rules for Low-Profit Products:")
print(low_profit_rules.head())

# OPTIONAL: Visualization of results (if needed)
# You can visualize the support, confidence, and lift of the filtered rules to understand relationships
plt.figure(figsize=(10, 6))
sns.scatterplot(x='support', y='confidence', size='lift', hue='lift', data=rules, palette='coolwarm')
plt.title('Association Rules: Support vs Confidence (Size=Lift)')
plt.show()
```
